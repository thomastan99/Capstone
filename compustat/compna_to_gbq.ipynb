```python
import pandas as pd
from google.oauth2 import service_account
import pandas_gbq as gbq
```

```python
credentials_path = 'token.json'

# Authenticate with your credentials
credentials = service_account.Credentials.from_service_account_file(
    credentials_path, scopes=['https://www.googleapis.com/auth/bigquery'])

# Set the credentials for pandas_gbq
gbq.context.credentials = credentials

project_id = 'capstone-398012'
dataset_id = 'capstone'
table_id = "comp_na_daily_all"
```

```python
df = pd.read_csv('comp_na_daily_all.csv')
df = df.astype(str)
```

```python
df.to_gbq(destination_table=f'{project_id}.{dataset_id}.compna_table', project_id=project_id, if_exists='replace')
```
