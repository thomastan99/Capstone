```python
import pandas as pd
from google.oauth2 import service_account
import pandas_gbq as gbq
```

```python
file_path = '../dataset/comp_na_daily_all.csv'

# Read the CSV file into a pandas DataFrame
chunks = pd.read_csv(file_path, chunksize=100000)
df = pd.concat(chunks)

df.head()
```

```python
df.describe()
```

```python
# drop cols w lots of missing value (threshold such that cutting down to 50%)

na_count_per_column = df.isna().sum()
half_length = len(df) / 2
# na_count_is_zero = na_count_per_column[na_count_per_column == 0]
na_count_50 = na_count_per_column[na_count_per_column < half_length].index

# print(na_count_50)

new_df = df[na_count_50]

# 2 letter cols are more impt - append Q ?
```

```python
new_df = new_df.astype(str)
new_df.head()
```

```python
credentials_path = 'token.json'

# Authenticate with your credentials
credentials = service_account.Credentials.from_service_account_file(
    credentials_path, scopes=['https://www.googleapis.com/auth/bigquery'])

# Set the credentials for pandas_gbq
gbq.context.credentials = credentials
project_id = 'capstone-402105'
dataset_id = 'capstone'
table_id = "compna_removed_na"
```

```python
new_df.to_gbq(destination_table=f'{project_id}.{dataset_id}.compna_removed_na', project_id=project_id, if_exists='replace')
```

```python
tickers = df['tic']
date = df['datadate']
data = df.fillna(0)
data_wo_string = data.drop(columns=data.select_dtypes(include='object'))
x_features = data_wo_string.drop(['gvkey', 'fyearq', 'fqtr', 'fyr', 'fyrc'], axis=1)
x_features.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(x_features)
X = scalar.transform(x_features)

from sklearn.decomposition import PCA
pca = PCA(n_components=3)
x_pca = pca.fit_transform(X)

combined_df = pd.concat([tickers, date, pd.DataFrame(x_pca)], axis=1)
combined_df.columns = ['Ticker', 'date', 'pc1', 'pc2', 'pc3']

combined_df.head(10)
```

```python
# Split the Date column into yyyy and mm columns
combined_df[['mm', 'dd', 'yyyy']] = combined_df['date'].str.split('/', expand=True)

# Convert the yyyy and mm columns to integers (optional)
combined_df['yyyy'] = combined_df['yyyy'].astype(int)
combined_df['mm'] = combined_df['mm'].astype(str)
combined_df = combined_df.drop(['dd'], axis=1)

combined_df.head(10)
```

```python
credentials_path = 'token.json'

# Authenticate with your credentials
credentials = service_account.Credentials.from_service_account_file(
    credentials_path, scopes=['https://www.googleapis.com/auth/bigquery'])

# Set the credentials for pandas_gbq
gbq.context.credentials = credentials

project_id = 'capstone-398012'
dataset_id = 'capstone'
table_id = "ticker_table"

from pandas_gbq import read_gbq

query = f"""
SELECT *
FROM `{project_id}.{dataset_id}.{table_id}`

"""

# Authenticate and read data from BigQuery into a DataFrame
ticker_df = read_gbq(query, project_id=project_id, dialect='standard')
ticker_df.head()
```

```python
ticker_df_test = ticker_df.head(100000)
combined_df_test = combined_df.head(100000)
```

```python
#merge both dfs based on ticker, yyyy, mm
merged_df = pd.merge(
    ticker_df_test,
    combined_df_test,
    on=['Ticker', 'yyyy', 'mm'],
    how='left'
)
merged_df
```
