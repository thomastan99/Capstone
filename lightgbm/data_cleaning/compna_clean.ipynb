```python
from google.oauth2 import service_account
import pandas as pd
import pandas_gbq as gbq
credentials_path = 'token.json'

# Authenticate with your credentials
credentials = service_account.Credentials.from_service_account_file(
    credentials_path, scopes=['https://www.googleapis.com/auth/bigquery'])

# Set the credentials for pandas_gbq
gbq.context.credentials = credentials
```

```python
project_id = 'capstone-398012'
dataset_id = 'capstone'
table_id = "compna_removed_na"
```

```python
from pandas_gbq import read_gbq

query = f"""
SELECT *
FROM `{project_id}.{dataset_id}.{table_id}`

"""

# Authenticate and read data from BigQuery into a DataFrame
df = read_gbq(query, project_id=project_id, dialect='standard')
df
```

```python
df.columns
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
cols = ['gvkey', 'datadate', 'sic', 'atq', 'ltq', 'dlcq', 'txditcq', 'cheq', 'lctq', 'actq', 'invtq', 'intanq', 'oibdpq', 'txpq', 'piq', 'niq', 'revtq', 'cogsq', 'oancfy', 'ivncfy', 'fincfy', 'dltisy', 'dltry', 'ceqq', 'dpq', 'saleq', 'saley', 'invchy', 'rectq', 'wcapq', 'prccq', 'epsf12', 'cshoq', 'dvy', 'req', 'dlttq', 'chechy']
print(len(cols))
```

```python
#['tieq', 'uopiq', 'uaptq', 'uceqq', 'uniamiq'] not in df
new2_df = df[cols]
new2_df
```

```python
#convert type to num
cols_convert = ['atq', 'ltq', 'dlcq', 'txditcq', 'cheq', 'lctq', 'actq', 'invtq', 'intanq', 'oibdpq', 'txpq', 'piq', 'niq', 'revtq', 'cogsq', 'oancfy', 'ivncfy', 'fincfy', 'dltisy', 'dltry', 'ceqq', 'dpq', 'saleq', 'saley', 'invchy', 'rectq', 'wcapq', 'prccq', 'epsf12', 'cshoq', 'dvy', 'req', 'dlttq', 'chechy']
new2_df[cols_convert] = new2_df[cols_convert].astype(float)
new2_df.dtypes
```

```python
new2_df.isna().sum().sort_values(ascending=False)
```

```python
new2_df.groupby(['gvkey']).apply(lambda x: x.isna().sum().sum()).sort_values(ascending=False)
```

```python
#create sector industry column
new2_df['sector_industry'] = new2_df['sic'].apply(lambda x: x[:2])
new2_df['sector_industry']
```

```python
#impute with sector median
grouped_df = new2_df.groupby('sector_industry')
grouped_df[cols_convert].median()
```

```python
median_values = grouped_df[cols_convert].transform('median')

new2_df[cols_convert] = new2_df[cols_convert].fillna(median_values)

new2_df
```

```python
new2_df.isna().sum().sort_values(ascending=False)
```

```python
new_df = new2_df
new_df
```

```python
new_df.isna().sum().sort_values(ascending=False)
```

```python
new_df['datadate'] = pd.to_datetime(new_df['datadate'])
new_df['datadate']
```

```python
#split to yyyy and mm column
new_df['yyyy'] = new_df['datadate'].dt.year
new_df['mm'] = new_df['datadate'].dt.month
new_df
```

```python
def expand_dates(ser):
    min_date = ser['datadate'].min()
    max_date = ser['datadate'].max()
    return pd.DataFrame({'datadate': pd.date_range(min_date, max_date, freq='M')})

# Apply the function within each 'id' group
newdf = new_df.groupby(['gvkey', 'yyyy']).apply(expand_dates).reset_index()

# Merge with the original DataFrame and forward fill missing values for the entire row
result = newdf.merge(new_df, on=['gvkey', 'datadate'], how='left').fillna(method='ffill')

print(result)
```

```python
result['mm'] = result['datadate'].dt.month
result
```

```python
new_df = result.drop(['yyyy_x', 'level_2'], axis=1)
new_df
```

```python
#feature engineer financial ratios

#create `current_ratio`
new_df['current_ratio'] = new_df['actq']/new_df['lctq']

#create `quick_ratio`
new_df['quick_ratio'] = (new_df['actq'] - new_df['invtq'])/new_df['lctq']

#create `cash_ratio`
new_df['cash_ratio'] = new_df['chechy']/new_df['lctq']

#create `net_working_capital`
new_df['net_working_capital'] = new_df['actq'] - new_df['lctq']

#create `debt_ratio`
new_df['debt_ratio'] = new_df['dlttq']/new_df['actq']

#create `debt_to_equity_ratio`
new_df['debt_to_equity_ratio'] = new_df['dlttq']/new_df['ceqq']

#create `equity_ratio`
new_df['equity_ratio'] = new_df['ceqq']/new_df['actq']

#create `financial_leverage_ratio`
new_df['financial_leverage_ratio'] = new_df['actq']/new_df['ceqq']

#create `cashflow_to_debt_ratio`
new_df['cashflow_to_debt_ratio'] = new_df['oancfy']/new_df['dlttq']

#create `net_profit_margin`
new_df['net_profit_margin'] = (new_df['revtq'] - new_df['cogsq'])/new_df['revtq']

#create `asset_turnover`
new_df['asset_turnover'] = new_df['saleq']/new_df['saley']

#create `receivables_turnover`
new_df['receivables_turnover'] = new_df['saleq']/new_df['rectq']

#create `day_sales_outstanding`
new_df['day_sales_outstanding'] = new_df['rectq']/new_df['revtq']

#create `working_capital_turnover`
new_df['working_capital_turnover'] = new_df['saleq']/new_df['wcapq']

#create `price_to_earnings`
new_df['price_to_earnings'] = new_df['prccq']/new_df['epsf12']

#create `retention_ratio`
new_df['retention_ratio'] = new_df['req']/new_df['niq']





```
