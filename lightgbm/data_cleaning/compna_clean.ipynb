```python
#define function to get average between current and past quarter 
def get_average(row):
    row['average_asset'] = (row['atq'] + row['atq'].shift(1)) / 2
    row['average_inventory'] = (row['invtq'] + row['invtq'].shift(1)) / 2
    row['average_receivables'] = (row['rectq'] + row['rectq'].shift(1)) / 2
    row['average_workingcapital'] = (row['wcapq'] + row['wcapq'].shift(1)) / 2
    return row
```

```python
#calculate average total assets, inventory, receivables and working capital
new_df = new_df.groupby(['gvkey']).apply(get_average)
new_df
```

```python
new_df.reset_index(drop=True, inplace=True)
```

```python
new_df
```

```python
#feature engineering of financial ratios

#create `current_ratio`
new_df['current_ratio'] = new_df['actq']/new_df['lctq']

#create `quick_ratio`
new_df['quick_ratio'] = (new_df['actq'] - new_df['invtq'])/new_df['lctq']

#create `cash_ratio`
new_df['cash_ratio'] = new_df['chechy']/new_df['lctq']

#create `net_working_capital`
new_df['net_working_capital'] = new_df['actq'] - new_df['lctq']

#create `debt_ratio`
new_df['debt_ratio'] = (new_df['dlttq'] + new_df['dlcq'])/new_df['atq']

#create `debt_to_equity_ratio`
new_df['debt_to_equity_ratio'] = (new_df['dlttq'] + new_df['dlcq'])/new_df['ceqq']

#create `equity_ratio`
new_df['equity_ratio'] = new_df['ceqq']/new_df['atq']

#create `cashflow_to_debt_ratio`
#debt is long-term + short-term debt
new_df['cashflow_to_debt_ratio'] = new_df['oancfy']/(new_df['dlttq'] + new_df['dlcq'])

#create `net_profit_margin`
#use niq as net profit
new_df['net_profit_margin'] = (new_df['niq']/new_df['revtq']) * 100

#create `return_on_assets`
new_df['return_on_assets'] = (new_df['niq']/new_df['atq']) * 100

#create `asset_turnover`
new_df['asset_turnover'] = new_df['revtq']/new_df['average_asset']

#create `inventory_turnover`
new_df['inventory_turnover'] = new_df['cogsq']/new_df['average_inventory']

#create `days_in_inventory`
new_df['days_in_inventory'] = 365/(new_df['cogsq']/new_df['average_inventory'])

#create `receivables_turnover`
new_df['receivables_turnover'] = new_df['revtq']/new_df['average_receivables']

#create `day_sales_outstanding`
new_df['day_sales_outstanding'] = 365/(new_df['revtq']/new_df['average_receivables'])

#create `working_capital_turnover`
new_df['working_capital_turnover'] = new_df['revtq']/new_df['average_workingcapital']

#create `price_to_earnings`
new_df['price_to_earnings'] = new_df['prccq']/new_df['epsfxq']

#create `retention_ratio`
new_df['retention_ratio'] = 1 - (new_df['dvpspq']/new_df['epsfxq'])

#create `gross_margin_ratio`
new_df['gross_margin_ratio'] = ((new_df['revtq'] - new_df['cogsq'])/new_df['revtq']) * 100

#create `operating_profit_margin`
new_df['operating_profit_margin'] = ((new_df['revtq'] - new_df['cogsq'] - new_df['xoprq'])/new_df['revtq']) * 100

#create `ebitda_margin` 
new_df['ebitda_margin'] = ((new_df['revtq'] - new_df['cogsq'] - new_df['xoprq'] + new_df['dpq'])/new_df['revtq']) * 100

#create `debt_service_coverage_ratio`
new_df['debt_service_coverage_ratio'] = (new_df['revtq'] - new_df['cogsq'] - new_df['xoprq'])/(new_df['dlttq'] + new_df['dlcq'])

#create `interest_coverage_ratio`
new_df['interest_coverage_ratio'] = new_df['oibdpq']/new_df['xintq']

#create `dividend_payout_ratio`
new_df['dividend_payout_ratio'] = (new_df['dvpspq']/new_df['epsfxq'])
```

```python
#list of engineered features
financial_ratios = [
    'current_ratio',
    'quick_ratio',
    'cash_ratio',
    'net_working_capital',
    'debt_ratio',
    'debt_to_equity_ratio',
    'equity_ratio',
    'cashflow_to_debt_ratio',
    'net_profit_margin',
    'return_on_assets',
    'asset_turnover',
    'inventory_turnover',
    'days_in_inventory',
    'receivables_turnover',
    'day_sales_outstanding',
    'working_capital_turnover',
    'price_to_earnings',
    'retention_ratio',
    'gross_margin_ratio',
    'operating_profit_margin',
    'ebitda_margin',
    'debt_service_coverage_ratio',
    'interest_coverage_ratio',
    'dividend_payout_ratio'
]
```

```python
imputed_df = new_df.copy()
imputed_df
```

```python
import warnings

# Suppress the "Mean of empty slice" warning
warnings.filterwarnings("ignore", category=RuntimeWarning)

for col in financial_ratios:
    imputed_df[col] = imputed_df.groupby(['sector_industry', 'quarter'])[col].transform(lambda x: x.fillna(x.median()))

# Reset the DataFrame index if needed
imputed_df.reset_index(drop=True, inplace=True)
```

```python
#use binary to create column to identify inf values
for col in undefined:
    imputed_df[f'{col}_inf'] = imputed_df[col].isin([np.inf, -np.inf]).astype(int)
```

```python
#expand dataset from quarterly to monthly
def expand_dates(ser):
    min_date = ser['datadate'].min()
    max_date = ser['datadate'].max()
    return pd.DataFrame({'datadate': pd.date_range(min_date, max_date, freq='M')})

# Apply the function within each 'id' group
ffdf = imputed_df.groupby(['gvkey']).apply(expand_dates).reset_index()

# Merge with the original DataFrame and forward fill missing values for the entire row
final_df = ffdf.merge(imputed_df, on=['gvkey', 'datadate'], how='left').fillna(method='ffill')

final_df
```

```python
#drop redundant columns
final_df = final_df.drop(['level_1'], axis=1)
final_df
```

```python
#create yyyy and mm columns
final_df['yyyy'] = final_df['datadate'].dt.year
final_df['mm'] = final_df['datadate'].dt.month
final_df
```
