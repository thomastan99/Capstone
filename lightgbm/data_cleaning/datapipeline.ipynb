```python
#winsorization function
import numpy as np

def winsorize(data, lower_percentile=0.001, upper_percentile=0.999):
    lower_threshold = np.percentile(data, lower_percentile * 100)
    upper_threshold = np.percentile(data, upper_percentile * 100)

    winsorized_data = np.where(data < lower_threshold, lower_threshold, data)
    winsorized_data = np.where(winsorized_data > upper_threshold, upper_threshold, winsorized_data)

    return winsorized_data
```
