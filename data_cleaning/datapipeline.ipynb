```python
from google.oauth2 import service_account
import pandas as pd
import pandas_gbq as gbq
credentials_path = '../token.json'

# Authenticate with your credentials
credentials = service_account.Credentials.from_service_account_file(
    credentials_path, scopes=['https://www.googleapis.com/auth/bigquery'])

# Set the credentials for pandas_gbq
gbq.context.credentials = credentials
```

```python
project_id = 'capstone-398012'
dataset_id = 'capstone'
table_id = "X_Y_merged"
```

```python
from pandas_gbq import read_gbq

query = f"""
SELECT *
FROM `{project_id}.{dataset_id}.{table_id}`

"""

# Authenticate and read data from BigQuery into a DataFrame
df = read_gbq(query, project_id=project_id, dialect='standard')
df
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
new_df = df
new_df
```

```python
nan_count = new_df.isna().sum(axis=1)

# Create a new column with the NaN count
new_df['nan_count'] = nan_count
new_df
```

```python
for i in range(22):
    print(str(i) + ":")
    print(len(new_df[new_df['nan_count']>i]))
```

```python
#drop all rows with nan_count >= 5

new_df.drop(new_df[new_df['nan_count'] >= 5].index, inplace=True)
new_df
```

```python
new_df.isna().sum().sort_values(ascending=False)
```

```python
#list of columns that contain NA 
na_cols = ['liqnonfinlevel', 'liqnonfintrend', 'dtdlevel', 'dtdtrend', 'DTDmedianNonFin', 'Sector_Number', 'DTDmedianFin', 'sigma', 'StkIndx']

#create new columns (_notNA) for each column that contains na values where 1 if original column is not na and 0 otherwise
for col in na_cols:
    new_col = col + '_notNA'
    new_df[new_col] = new_df[col].notna().astype(int)

new_df

```

```python
new_df.shape
```

```python
grouped_df = new_df.groupby('CompNo')
grouped_df[na_cols].mean()
```

```python
mean_values = grouped_df[na_cols].transform('mean')

new_df[na_cols] = new_df[na_cols].fillna(mean_values)

new_df
```

```python
print(new_df[new_df.isna().any(axis=1)])
```

```python
new_df.isna().sum().sort_values(ascending=False)
```

```python
new_df.to_gbq(destination_table=f'{project_id}.{dataset_id}.filled_table', project_id=project_id, if_exists='replace')
```
